```python
def create_splits(scan_df,k,seed=100):
#     KFold(n_splits=5,shuffle=True, random_state=seed)

    pre_df=scan_df[scan_df['Pre']==1]
    post_df=scan_df[scan_df['Pre']==0]
    patients=list(scan_df['ID'].unique())
    train_pats,test_pats=train_test_split(patients)
    
    
    
    train_pats=patients
    stat_cols=[c for c in scan_df if 'adprob' in c]
    meta_cols=['PreID','PostID','Same']
    data=[]
    
    kf = KFold(n_splits=k)
    for split, (train_index, test_index) in enumerate(kf.split(patients)):
        for i in test_index:
            ti=patients[i]
            print(ti,'TI')
            t_pre=pre_df[pre_df['ID']==ti]
            t_pre_stat=t_pre[stat_cols]
            
            for j in test_index:
                tj=patients[j]
                t_post=pre_df[pre_df['ID']==tj]
                t_post_stat=t_post[stat_cols]
                difference=np.array(t_post_stat)-np.array(t_pre_stat)

#                 print(t_pre_stat,'pre')
#                 print(t_post_stat,'post')
                label=1 if tj==ti else 0
                meta=[ti,tj,label,split]
                row=np.concatenate(meta,difference)
                print(row.shape,'row shape')
                difference[['PatPre','PatPost','label','split']]=meta
                data.append(difference)
    print(data)
    data=pd.DataFrame(data)
            
    print(data,'DATA SHAPE')
    
        
            
            
            
```
